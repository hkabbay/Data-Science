```python
# Create dataset
import pandas as pd
import numpy as np
input = np.array([[0,0] ,[0,1] ,[1,0] ,[1,1]])
target = np.array([[1], [0], [0], [1]])
df = pd.DataFrame(input, columns=['x', 'y'])
df['x XNOR y'] = target
df
```

```python
# Import required libraries
from keras import models
from keras import layers
```

```python
# Build model
network = models.Sequential()
network.add(layers.Dense(2, activation='relu', input_dim=2))
# network.add(layers.Dense(1, activation='relu'))
network.add(layers.Dense(1, activation='sigmoid'))
network.compile(optimizer='RMSprop',
                loss='binary_crossentropy',
                metrics=['binary_accuracy'])
```

```python
network.fit(input, target, epochs=1000)
```

```python
scores = network.evaluate(input, target)
print("\n%s: %.2f%%" % (network.metrics_names[1], scores[1]*100))
```

```python
print (network.predict(input).round())
print ("weights", network.layers[0].get_weights()[0])
print ("biases", network.layers[0].get_weights()[1])
```
