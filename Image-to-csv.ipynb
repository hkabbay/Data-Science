```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = files.upload()
```

```python
mycolor = cv2.imread('cat_baby01.jpg')
resized = cv2.resize(mycolor, (128,128), interpolation = cv2.INTER_AREA)
mygray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
```

```python
print(mygray.shape)
plt.imshow(mygray, cmap='gray')
plt.colorbar()
plt.show()
```

```python
np.savetxt('my-cat.csv', mygray, fmt='%.1f', delimiter=',')
files.download('my-cat.csv')
```
